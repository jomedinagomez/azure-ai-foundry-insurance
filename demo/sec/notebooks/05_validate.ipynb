# 05 — Validate against expected output

Compares a `SecExtractionResult` against `demo/sec/reference/expected-output/<stem>.json`.
Drives the **Validation** panel on the SEC page and the `/sec/validate`
endpoint via `sec_service.validate(result)`.

Expected-output files are line-item lists per statement type; the
validator reports per-statement matched / missing / extra row counts and
an overall match rate.


```python
from _lib import sec_service
sample = sec_service.list_samples()[0]['file_name']
result = sec_service.run_extraction(sample, use_cache=True)
print('statements:', [s['statement_type'] for s in result['statements']])
print('rows per statement:', {s['statement_type']: len(s['rows']) for s in result['statements']})
```

```python
report = sec_service.validate(result)
print('has_ground_truth:', report['has_ground_truth'])
print('overall match rate:', report['overall_match_rate'])
for s in report['statements']:
    print(f"  {s['statement_type']:22s}  expected={s['expected_rows']:3d}  actual={s['actual_rows']:3d}  matched={s['matched_rows']:3d}  missing={len(s['missing_rows']):3d}  extra={len(s['extra_rows']):3d}")
```
