# 01 — Deploy SEC analyzers to Content Understanding

Loads the classifier + analyzer JSON templates from
`demo/sec/reference/analyzer-templates/` and registers them with
Content Understanding via `client.begin_create_analyzer(..., allow_replace=True)`.

This is exactly what the API does implicitly on first run via
`sec_service.ensure_analyzers()` — the SEC page just hides it behind
the **Run extraction** button.


```python
from _lib import sec_service
print('Classifier ID:', sec_service.CLASSIFIER_ID)
print('Analyzer ID:  ', sec_service.ANALYZER_ID)
print('Templates in :', sec_service.TEMPLATE_DIR)
```

## Inspect the templates

```python
import json
classifier_tmpl = sec_service._classifier_template()
analyzer_tmpl = sec_service._analyzer_template()
print('Classifier categories:', list(classifier_tmpl['config']['contentCategories'].keys()))
print('Analyzer fields      :', list(analyzer_tmpl['fieldSchema']['fields'].keys()))
```

## Deploy (idempotent)

Safe to re-run. `ensure_analyzers()` skips deployment when an analyzer of the same ID already exists; pass `force_replace=True` to redeploy.

```python
statuses = sec_service.ensure_analyzers()
for analyzer_id, status in statuses.items():
    print(f'  {analyzer_id}: {status}')
```

**Parity check** — this is the exact call the API's `/sec/extract` endpoint makes when a sample is run without a cached result.
