# 04 — End to end

Single cell that deploys analyzers (if needed), iterates through every bundled sample, runs the matching scenario analyzer, and prints a concise verdict. Useful for smoke-testing a fresh environment.

```python
from _lib import pro_service

print('Deploying analyzers (no-op if present)...')
print(pro_service.deploy_all(overwrite=False))
print()

for m in pro_service.list_sample_manifests():
    print(f'== {m.id} ({m.scenario}) ==')
    _, files = pro_service.load_sample(m.id)
    if m.scenario == 'claims':
        r = pro_service.analyze_claims(files, sample_id=m.id)
        print(f'  coverage  = {r.fields.coverage_applies}')
        print(f'  docset    = {r.fields.document_set_completeness}')
        print(f'  estimate  = {r.fields.estimated_total}')
    else:
        r = pro_service.analyze_fraud(files, sample_id=m.id)
        print(f'  score     = {r.risk_score}/100  band={r.risk_band}')
        print(f'  signals   = {len(r.rule_signals)} rule + {len(r.cu_signals)} CU')
    print()
```
